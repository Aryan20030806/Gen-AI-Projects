**FINE TUNNING A DATASET**

**STEP 1 : INSTALLING LIB**

```python
# installing all the req lib
!pip install datasets
```

**STEP 2 : LOADING THE DATASET**

```python
# loading the dataset
from datasets import load_dataset
dataset = load_dataset("stanfordnlp/imdb")

```

```python
# viewing the dataset
dataset
```

**STEP 3 : PREPROCESSING**

```python
# perform preprocessing and tokenization

from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

# tokenize the dataset
def tokenize_function(examples):
    return tokenizer(examples["text"], padding="max_length", truncation=True)

tokenized_datasets = dataset.map(tokenize_function, batched=True)

```

```python
# view tokenized dataset
tokenized_datasets
```

**STEP 4 : SETTING UP TRAINING ARGUMENTS**

```python
from transformers import TrainingArguments
training_args = TrainingArguments(
    output_dir = './results',         # output directory (storing of result in folder)
    eval_strategy='epoch',            # evaluate every epoch
    learning_rate=2e-5,               # learning rat
    per_device_train_batch_size = 16, # batch size of traning
    per_device_eval_batch_size = 16,  # batch size of evaluation
    num_train_epochs = 1,             # number of training epoch
    weight_decay=0.01,                # strength of weight decay
    report_to = 'none'                # disable wandb reporting
)
training_args
```

**STEP 5 : INITIALIZE THE MODEL**

```python
from transformers import AutoModelForSequenceClassification, Trainer

# load the pretrained model
model = AutoModelForSequenceClassification.from_pretrained("bert-base-uncased", num_labels=2)

# initialize the trainer
trainer = Trainer(
     model=model,
     args=training_args,
     train_dataset=tokenized_datasets["train"],
     eval_dataset=tokenized_datasets["test"],
 )
```

**TRAINING ThE MODEL**

```python
# training the model
trainer.train()
```

STEP 6 **EVALUATE THE MODEL**

```python
result = trainer.evaluate()
print(result)
```

**STEP 7 : SAVING THE MODEL FOR LATER USE**

```python
model.save_pretrained("./fine-tuned-model")
tokenizer.save_pretrained("./fine-tuned-model")
```
