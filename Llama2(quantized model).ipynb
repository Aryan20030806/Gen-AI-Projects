```python
model_name = "TheBloke/Llama-2-13B-chat-GGML"
model_basename = "llama-2-13b-chat.ggmlv3.q5_0.bin"
```

```python
from huggingface_hub import hf_hub_download
from llama_cpp import Llama
```

**2) downloading the model**

```python
model_path = hf_hub_download(repo_id=model_name,filename=model_basename)
```

```python
model_path
```

**3) LOADING THE MODEL**

```python
lcpp_llm = None
lcpp_llm = Llama(
    model_path=model_path,
    n_threads=2, # no of cpu cores to be used
    n_batch=512, # dependend upon gpu and vram
    n_gpu_layers=32 # based upon gpu and vram
)
```

```python
lcpp_llm.params.n_gpu_layers # to check the no of gpu layers in the gpu
```

**4) creating a prompt template**

```python
# TEMPLATE FORM HF WE CAN USE ANY SORT OF INPUT WE REQ
prompt = "write a code in python for adding two numbers"
prompt_template = f"""SYSTEM : YOU ARE A HELPFULL AI ASSISTANT MADE BY ARYAN SINGH.

USER : {prompt}

ASSISTANT:
"""
```

**5) CREATING THE RESPONSE**

```python
response = lcpp_llm(prompt=prompt_template,max_tokens=256,temperature=0.5,
                    top_p=0.95,repeat_penalty=1.2,
                    top_k=150,echo=True)
```
