```python
from langchain import PromptTemplate, HuggingFaceHub, LLMChain
from langchain_huggingface import HuggingFaceEndpoint
import os
```

**2) SETTING UP ENV**

```python
os.environ['HUGGINGFACEHUB_API_TOKEN'] = ""
```

**APPROACH 1** : ACCESS THE MODELS THROUGH API

```python
# setting the prompt tempelate
prompt = PromptTemplate(
    input_variables=["product"],
    template="what is a good name for a company tht makes {product}"
)
```

```python
# defining the model
llm_model = HuggingFaceHub(
    repo_id="google/flan-t5-large",
    task="text2text-generation",
    huggingfacehub_api_token=os.environ["HUGGINGFACEHUB_API_TOKEN"],
    model_kwargs={"temperature": 0, "max_new_tokens": 64}
)
```

```python
# INITIALIZING THE CHAIN
chain = LLMChain(llm=llm_model, prompt=prompt)
response = chain.invoke({"product": "skincare products"})
print(response["text"])
```

**APPROACH 2** : DOWNLOADING THE MODEL LOCALLY

```python
from langchain.llms import HuggingFacePipeline
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline, AutoModelForSeq2SeqLM
```

```python
model_id = "google/flan-t5-large"
tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForSeq2SeqLM.from_pretrained(model_id,device_map='auto')
```

```python
pipeline = pipeline(
    "text2text-generation",
    model=model,
    tokenizer=tokenizer,
    max_length=128
)
```

```python
local_llm= HuggingFacePipeline(pipeline=pipeline)
```

```python
# setting the prompt tempelate
prompt = PromptTemplate(
    input_variables=["product"],
    template="what is a good name for a company tht makes {product}"
)
```

```python
# INITIALIZING THE CHAIN
chain = LLMChain(llm=llm_model, prompt=prompt)
response = chain.invoke({"product": "skincare products"})
print(response["text"])
```
